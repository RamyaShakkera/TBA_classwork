## Count of specified columns

```python
from collections import Counter
import pandas as pd
import regex as re

# Define the tokenize function
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

# Specify the file path
file_path = "C:/Users/shakkerar/Desktop/Week2_textbased_analysis/un-general-debates-blueprint.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Assuming 'text' column exists in your CSV file
# If not, replace 'text' with the actual column name containing the text
df['tokens'] = df['text'].apply(lambda x: tokenize(str(x)))

# Use Counter to count the occurrences of specific tokens in the entire dataset
counter = Counter()
df['tokens'].map(counter.update)

# Display the count of specific tokens
tokens_to_count = ["nation", "united", "international", "world", "countries"]
for token in tokens_to_count:
    print(f"{token}: {counter[token]}")
```

## Frequency

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)

```

```python
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
