```python
import pandas as pd
import regex as re

# Define the pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Define the tokenize function
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

# Define the remove_stop function
def remove_stop(tokens):
  
    stopwords = ["the", "and", "of", "in", "to", "a", "is", "it"]
    return [t for t in tokens if t.lower() not in stopwords]

# Define the prepare function
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens

# Specify the file path for Titanic dataset
file_path = "C:/Users/shakkerar/Desktop/Semester2_content/TBA/Titanic-Dataset.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Apply the pipeline to the 'Name' column
df['tokens'] = df['Name'].apply(lambda x: prepare(str(x), pipeline=pipeline))

# Display the DataFrame with the processed tokens
print(df[['Name', 'tokens']].head())
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image

file_path = "C:/Users/shakkerar/Desktop/Semester2_content/TBA/Titanic-Dataset.csv"
df = pd.read_csv(file_path)

# Concatenate text values from all relevant columns
text = ' '.join(df.astype(str).values.flatten())

# Define stopwords to exclude common words
stopwords = ["the", "and", "of", "in", "to", "a", "is", "it"]

# Load the ship-shaped mask image
ship_mask = np.array(Image.open("C:/Users/shakkerar/Desktop/titanic.webp"))

# Create WordCloud with adjusted parameters
plt.figure(figsize=(14, 7))
wc = WordCloud(max_words=100, stopwords=stopwords, background_color='white',
               contour_width=0, contour_color='black', colormap='viridis',
               mask=ship_mask)
wc.generate(text)

# Display WordCloud in the shape of a ship
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
from io import BytesIO
import requests
import pandas as pd

# Assuming you have loaded the Titanic dataset into 'df'
file_path = "C:/Users/shakkerar/Desktop/Semester2_content/TBA/Titanic-Dataset.csv"
df = pd.read_csv(file_path)

# Concatenate text values from all relevant columns
text = ' '.join(df.astype(str).values.flatten())

# From URL
response = requests.get("https://365psd.com/images/istock/previews/1004/100424327-vector-cloud-icon.jpg")
mask = np.array(Image.open(BytesIO(response.content)))

# Colors
image_colors = ImageColorGenerator(mask)

wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.figure(figsize=(8, 8))
plt.axis("off")
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")

plt.show()
```
